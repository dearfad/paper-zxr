```python
import cv2
import numpy as np
```

```python
image_path = "../img/mcf-7-test.tif"
```

```python
image = cv2.imread(image_path)
```

```python
 # 亮度阈值
BRIGHTNESS_MEAN_MIN: float = 50.0
BRIGHTNESS_MEAN_MAX: float = 200.0
BRIGHTNESS_STD_MIN: float = 20.0
```

```python
# 1. 清晰度检测
# 清晰度阈值 (拉普拉斯方差)
LAPLACIAN_VAR_THRESHOLD: float = 50.0

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
# 拉普拉斯算子
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

variance = laplacian.var()

if variance < LAPLACIAN_VAR_THRESHOLD:
    print(f"清晰度不足 (拉普拉斯方差: {variance:.2f} < {LAPLACIAN_VAR_THRESHOLD})")
else:
    print(f"清晰度正常 (拉普拉斯方差: {variance:.2f} > {LAPLACIAN_VAR_THRESHOLD})")
```

```python
"""
计算亮度统计特征
Returns: {'mean': 均值, 'std': 标准差, 'min': 最小值, 'max': 最大值}
"""
# 亮度阈值
BRIGHTNESS_MEAN_MIN: float = 50.0
BRIGHTNESS_MEAN_MAX: float = 200.0
BRIGHTNESS_STD_MIN: float = 20.0

mean = np.mean(gray)
std = np.std(gray)
min = np.min(gray)
max = np.max(gray)

print ('mean:', mean, 'std:', std, 'min:', min, 'max:', max)
```

**🎯 质控标准（项目特定）**

| 指标         | MCF-7标准         | MDA-MB-231标准    | 说明                           |
| ------------ | ----------------- | ----------------- | ------------------------------ |
| **清晰度**   | 拉普拉斯方差 > 55 | 拉普拉斯方差 > 45 | MDA-MB-231边缘不规则，阈值放宽 |
| **亮度均值** | [50, 200]         | [40, 220]         | 避免过曝/欠曝                  |
| **对比度**   | 标准差 > 20       | 标准差 > 20       | 确保细胞边缘清晰               |
| **划痕**     | 无严重亮线        | 无严重亮线        | 霍夫直线检测                   |
| **气泡**     | < 3个             | < 3个             | 圆形度>0.7的亮斑               |

**表格项目**
- timestamp': 录入时间
- sample_id: 样本ID (如: MCF7_001)
- image_path: 图像完整路径
- cell_line: 细胞系类型 ('MCF7' 或 'MDA231')
- timepoint: 时间点 (如: 24h)
- view_id: 视野编号 (如: F01)
- reject_reason: 不合格原因
- severity: 严重程度
- operator: 操作员
- notes: 详细备注
- reviewed: 是否审核

**REJECT_REASONS**
- 图像模糊
- 曝光过度 
- 曝光不足
- 划痕干扰
- 气泡干扰
- 杂质污染
- 视野偏移
- 细胞脱落
- 培养失败形态
- 其他

```python
import cv2
import numpy as np
```

```python
image_path = "../img/24h-40x-center.tif"
# image_path ="../img/mcf-7-test.tif"
```

```python
 # 亮度阈值
BRIGHTNESS_MEAN_MIN: float = 50.0
BRIGHTNESS_MEAN_MAX: float = 200.0
BRIGHTNESS_STD_MIN: float = 20.0

# 读取图片
image = cv2.imread(image_path)

# 1. 清晰度检测
# 清晰度阈值 (拉普拉斯方差)
LAPLACIAN_VAR_THRESHOLD: float = 50.0

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
# 拉普拉斯算子
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

variance = laplacian.var()

if variance < LAPLACIAN_VAR_THRESHOLD:
    print(f"清晰度不足 (拉普拉斯方差: {variance:.2f} < {LAPLACIAN_VAR_THRESHOLD})")
else:
    print(f"清晰度正常 (拉普拉斯方差: {variance:.2f} > {LAPLACIAN_VAR_THRESHOLD})")
```

```python
"""
计算亮度统计特征
Returns: {'mean': 均值, 'std': 标准差, 'min': 最小值, 'max': 最大值}
"""
# 亮度阈值
BRIGHTNESS_MEAN_MIN: float = 50.0
BRIGHTNESS_MEAN_MAX: float = 200.0
BRIGHTNESS_STD_MIN: float = 20.0

mean = np.mean(gray)
std = np.std(gray)
min = np.min(gray)
max = np.max(gray)

print ('mean:', mean, 'std:', std, 'min:', min, 'max:', max)
```
